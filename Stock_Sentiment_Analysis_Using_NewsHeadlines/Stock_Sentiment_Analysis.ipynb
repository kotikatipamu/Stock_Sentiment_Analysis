```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Stock_Sentiment.csv')
```

```python
df.head()
```

```python
# Creating train and test dataset
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
train.shape
```

```python
test.shape
```

```python
# We to remove punctuations
data = train.iloc[:,2:]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
#Removing columns names for ease of access
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns = new_index
data.head()
```

```python
#converting the headlines in lowercase
for i in new_index:
    data[i] = data[i].str.lower()
data.head()
```

```python
#Combining the records(roww)
headlines = []
for row in range(0,len(data.index)):
    headlines.append(" ".join(str(x) for x in data.iloc[row,0:25]))
headlines[0]
```

```python
#We have to convert the words into vectors
# using countvectorizer--creating the Bag of words
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range = (2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# Same steps we have to follow to test dataset also
test_transform = []
for row in range(0,len(test.index)):
    test_transform.append(" ".join(str(x) for x in data.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
prediction = randomclassifier.predict(test_dataset)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix = confusion_matrix(test['Label'],prediction)
print(matrix)
report = classification_report(test['Label'],prediction)
print(report)
score = accuracy_score(test['Label'],prediction)
print(score)
```

```python
# Using TF-IFDvectorizer for better Accuracy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
tfidfvector = TfidfVectorizer(ngram_range = (2,2))
traindataset = tfidfvector.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# Same steps we have to follow to test dataset also
test_transform = []
for row in range(0,len(test.index)):
    test_transform.append(" ".join(str(x) for x in data.iloc[row,2:27]))
test_dataset = tfidfvector.transform(test_transform)
prediction = randomclassifier.predict(test_dataset)
```

```python
matrix = confusion_matrix(test['Label'],prediction)
print(matrix)
report = classification_report(test['Label'],prediction)
print(report)
score = accuracy_score(test['Label'],prediction)
print(score)
```

```python
naive_cls.fit(traindataset,train['Label'])
```
